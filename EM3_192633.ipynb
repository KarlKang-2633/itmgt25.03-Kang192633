### Mobile Number Cleaning

```python
# mobile number cleaning

# dump recweek data
recweek
```

```python
# write code here

# I used regex because I wanted an exact match

import re 

## define function 
def remove63_add0(phonenumber):
    return re.sub(r"\A9([1-2][7-8])", r"09\1", re.sub(r"\+63 ", r"", phonenumber)) 
# nested re.sub; uses the initial removal of +63 as source for substituting 9([1-2][7-8]) with 09\1
# \A is used to match only at the beginning of the string

## apply function 
recweek["phonenumber"] = recweek["phonenumber"].apply(remove63_add0)


# dump cleaned phonenumber recweek
recweek
```

#### Setting up new df to play with

```python
# write code here

# make copy of final_recweek
copy_finrecweek = final_recweek.copy()

# adding new columns with 
copy_finrecweek["Smart"] = copy_finrecweek["phonenumber"].str.contains(r"0918 ")
copy_finrecweek["Globe"] = copy_finrecweek["phonenumber"].str.contains(r"0917 |0927 ")
# add empty column "carrier", where we will put whether the individual's phone number is Smart or Globe
copy_finrecweek["carrier"] = ""

# locate where Smart/Globe is true, then add "Smart"/"Globe" into carrier column
copy_finrecweek.loc[copy_finrecweek["Smart"] == True, "carrier"] = "Smart"
copy_finrecweek.loc[copy_finrecweek["Globe"] == True, "carrier"] = "Globe"

# dump
copy_finrecweek 
```

#### First Output

```python

# pass in a list of columns we want for multilevel filtering
# we only want to show one output for the count so we just choose any of the column to show 
# - in this case I just chose firstname for convience
copy_finrecweek.groupby(["carrier","gender"]).count().firstname
```

#### Second Output (dataframe)

```python

# assign a new variable carriergender_countdf as dataframe;  
# use groupby method, passing in a list as the argument so that we get sort of like two indices;
# we use method aggregate for convenience;
# - it automatically adds a count column, then reset index to clean the table
carriergender_countdf = copy_finrecweek.groupby(by = ["carrier","gender"]).firstname.agg(["count"]).reset_index()

# dump
carriergender_countdf
```
